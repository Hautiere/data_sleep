```python
# Cell 1: Import des bibliothèques nécessaires
import json
import pandas as pd
file_path = './2021-11-27_2022-03-07_102185832_sleepData.json'

# Cell 2: Charger le fichier JSON
def load_json_file(file_path):
    with open(file_path, 'r') as f:
        data = json.load(f)
    return data

# Cell 3: Afficher un extrait des données pour visualiser la structure
def display_json_sample(data, sample_size=2):
    return data[:sample_size]

# Cell 4: Convertir les données JSON en DataFrame pandas
def json_to_dataframe(data):
    df = pd.json_normalize(data)
    return df

# Cell 5: Analyser la structure des données et les premières lignes du DataFrame
def analyze_dataframe(df):
    print("Structure des données:")
    print(df.info())
    print("\nPremières lignes du DataFrame:")
    return df.head()

# Cell 6: Charger, visualiser et analyser le fichier JSON
# Remplacez 'votre_fichier.json' par le chemin réel vers votre fichier JSON
# file_path = 'votre_fichier.json'  # Mettre à jour avec le bon chemin
data = load_json_file(file_path)

# Afficher un extrait du contenu JSON
sample_data = display_json_sample(data)
print("Extrait des données JSON:")
print(sample_data)

# Convertir le JSON en DataFrame
df = json_to_dataframe(data)

# Analyser la structure du DataFrame
analyze_dataframe(df)
```

```python
# Cell 1: Import required libraries
import json
import datetime
from datetime import timedelta
file_path = './2021-11-27_2022-03-07_102185832_sleepData.json'

# Cell 2: Load JSON data from file
def load_sleep_data(file_path):
    with open(file_path, 'r') as f:
        sleep_data = json.load(f)
    return sleep_data

# Cell 3: Calculate total sleep time in hours
def calculate_total_sleep_time(data):
    total_sleep_seconds = data['deepSleepSeconds'] + data['lightSleepSeconds'] + data['remSleepSeconds']
    return total_sleep_seconds / 3600  # Convert seconds to hours

# Cell 4: Calculate sleep efficiency
def calculate_sleep_efficiency(data):
    sleep_duration = calculate_total_sleep_time(data)
    sleep_start = datetime.datetime.strptime(data['sleepStartTimestampGMT'], '%Y-%m-%dT%H:%M:%S.%f')
    sleep_end = datetime.datetime.strptime(data['sleepEndTimestampGMT'], '%Y-%m-%dT%H:%M:%S.%f')
    time_in_bed = (sleep_end - sleep_start).total_seconds() / 3600  # Convert to hours
    sleep_efficiency = (sleep_duration / time_in_bed) * 100
    return sleep_efficiency

# Cell 5: Generate a basic summary report
def generate_sleep_report(sleep_data):
    total_sleep_hours = []
    sleep_efficiencies = []
    
    for entry in sleep_data:
        total_sleep = calculate_total_sleep_time(entry)
        efficiency = calculate_sleep_efficiency(entry)
        
        total_sleep_hours.append(total_sleep)
        sleep_efficiencies.append(efficiency)
        
        print(f"Date: {entry['calendarDate']}")
        print(f"Total Sleep Time: {total_sleep:.2f} hours")
        print(f"Sleep Efficiency: {efficiency:.2f}%")
        print(f"Overall Sleep Score: {entry['sleepScores']['overallScore']}")
        print("-" * 40)
    
    avg_sleep_time = sum(total_sleep_hours) / len(total_sleep_hours)
    avg_efficiency = sum(sleep_efficiencies) / len(sleep_efficiencies)
    
    print("Summary Report")
    print("=" * 40)
    print(f"Average Sleep Time: {avg_sleep_time:.2f} hours")
    print(f"Average Sleep Efficiency: {avg_efficiency:.2f}%")
    print("=" * 40)

# Cell 6: Load data and generate report
# Replace 'your_file_path.json' with the actual path to your JSON file
# file_path = 'your_file_path.json'  # Update with the correct path
sleep_data = load_sleep_data(file_path)

# Generate the sleep report
generate_sleep_report(sleep_data)
```
