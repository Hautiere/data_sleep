```python
import os
from fitparse import FitFile
import matplotlib.pyplot as plt
import folium

def convert_semicircles_to_degrees(semicircles):
    return semicircles * (180 / 2**31)

def extract_gps_data(fitfile_path):
    fitfile = FitFile(fitfile_path)
    lats, longs = [], []

    # Iterate over all messages of type 'record'
    for record in fitfile.get_messages('record'):
        lat = None
        lon = None
        for record_data in record:
            if record_data.name == 'position_lat':
                lat = convert_semicircles_to_degrees(record_data.value)
            elif record_data.name == 'position_long':
                lon = convert_semicircles_to_degrees(record_data.value)
        if lat is not None and lon is not None:
            lats.append(lat)
            longs.append(lon)
    
    return lats, longs

def plot_course_on_map(latitudes, longitudes, output_html='course_map.html'):
    # Initialize the map centered around the first point
    course_map = folium.Map(location=[latitudes[0], longitudes[0]], zoom_start=13)

    # Combine latitudes and longitudes into pairs
    coordinates = list(zip(latitudes, longitudes))

    # Add the course to the map
    folium.PolyLine(locations=coordinates, color="blue", weight=2.5, opacity=1).add_to(course_map)

    # Save map to an HTML file
    course_map.save(output_html)
    print(f"Map has been saved to {output_html}")

def plot_course_on_graph(latitudes, longitudes):
    plt.figure(figsize=(10, 6))
    plt.plot(longitudes, latitudes, color='blue', marker='o', markersize=2)
    plt.title('Course Graph')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.grid(True)
    plt.show()

def main():
    fitfile_path = 'francois.hautiere@gmail.com_151427540440.fit'  # Replace with your FIT file path

    if not os.path.exists(fitfile_path):
        print(f"File not found: {fitfile_path}")
        return

    latitudes, longitudes = extract_gps_data(fitfile_path)

    if len(latitudes) == 0 or len(longitudes) == 0:
        print("No GPS data found in FIT file.")
        return

    # Plot the course on a graph
    plot_course_on_graph(latitudes, longitudes)

    # Plot the course on an interactive map
    plot_course_on_map(latitudes, longitudes)

if __name__ == "__main__":
    main()
```

```python
import os
from fitparse import FitFile
from datetime import datetime

def convert_semicircles_to_degrees(semicircles):
    return semicircles * (180 / 2**31)

def extract_gps_data(fitfile_path):
    fitfile = FitFile(fitfile_path)
    data = []

    for record in fitfile.get_messages('record'):
        record_data = {}
        for field in record:
            if field.name == 'position_lat':
                record_data['latitude'] = convert_semicircles_to_degrees(field.value)
            elif field.name == 'position_long':
                record_data['longitude'] = convert_semicircles_to_degrees(field.value)
            elif field.name == 'distance':
                record_data['distance'] = field.value / 1000  # Convert to kilometers
            elif field.name == 'speed':
                record_data['speed'] = field.value * 3.6  # Convert to km/h
            elif field.name == 'timestamp':
                record_data['timestamp'] = field.value

        if 'latitude' in record_data and 'longitude' in record_data:
            if 'speed' not in record_data:
                record_data['speed'] = None
            data.append(record_data)

    # Sort data by timestamp to ensure correct order
    data.sort(key=lambda x: x['timestamp'])
    return data
```

```python
import pandas as pd

def create_kilometer_matrix(data):
    # Create a DataFrame
    df = pd.DataFrame(data)

    # Convert timestamps to datetime
    df['timestamp'] = pd.to_datetime(df['timestamp'], unit='s')  # Assuming FIT file uses UNIX time
    
    # Calculate time difference in seconds
    df['time_diff'] = df['timestamp'].diff().dt.total_seconds().fillna(0)
    
    # Calculate distance difference in kilometers
    df['distance_diff'] = df['distance'].diff().fillna(0)
    
    # Calculate speed (km/h)
    df['speed'] = (df['distance_diff'] / df['time_diff']) * 3600  # Convert m/s to km/h
    
    # Fill NaNs for speed with previous value or 0 if no previous value
    df['speed'] = df['speed'].fillna(method='ffill').fillna(0)
    
    # Round the distance to the nearest whole number and group by this value
    df['distance_rounded'] = df['distance'].round(0)
    
    # Group by rounded distance and get the average speed and timestamp at each kilometer
    df_km = df.groupby('distance_rounded', as_index=False).agg({'speed': 'mean', 'timestamp': 'first'})

    # Rename columns for clarity
    df_km.rename(columns={'distance_rounded': 'distance_km'}, inplace=True)
    
    return df_km
```

```python
import pandas as pd

def create_kilometer_matrix(data):
    # Create a DataFrame
    df = pd.DataFrame(data)

    # Convert timestamps to datetime
    df['timestamp'] = pd.to_datetime(df['timestamp'], unit='s')  # Assuming FIT file uses UNIX time
    
    # Calculate time difference in seconds
    df['time_diff'] = df['timestamp'].diff().dt.total_seconds().fillna(0)
    
    # Calculate distance difference in kilometers
    df['distance_diff'] = df['distance'].diff().fillna(0)
    
    # Calculate speed (km/h)
    df['speed'] = (df['distance_diff'] / df['time_diff']) * 3600  # Convert m/s to km/h
    
    # Fill NaNs for speed with previous value or 0 if no previous value
    df['speed'] = df['speed'].fillna(method='ffill').fillna(0)
    
    # Round the distance to the nearest whole number and group by this value
    df['distance_rounded'] = df['distance'].round(0)
    
    # Group by rounded distance and get the average speed, altitude, and timestamp at each kilometer
    df_km = df.groupby('distance_rounded', as_index=False).agg({
        'speed': 'mean',
        'altitude': 'mean',
        'timestamp': 'first'
    })

    # Rename columns for clarity
    df_km.rename(columns={'distance_rounded': 'distance_km'}, inplace=True)
    
    return df_km
```
